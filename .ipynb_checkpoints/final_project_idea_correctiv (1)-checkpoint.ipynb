```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('messstellen.csv')
df
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
#handle null values
```

```python
# Check for NaN values
df.isna().sum()

# Fill NaN in numerical columns with the median
df['trend_normalized'].fillna(df['trend_normalized'].median(), inplace=True)
df['p_value'].fillna(df['p_value'].median(), inplace=True)

# Drop rows where 'trend_bin' is NaN since it's categorical and critical for our analysis
df.dropna(subset=['trend_bin'], inplace=True)

# Verify no critical NaNs remain
df.isna().sum()
```

### 	Identify locations classified as â€˜stark sinkendâ€™ or â€˜stark steigendâ€™ to focus on dramatic changes
###  Filter out the locations with strong increasing or decreasing trends based on trend_bin

```python
# Filter for strong changes
strong_changes = df[(df['trend_bin'] == 'stark sinkend') | (df['trend_bin'] == 'stark steigend')]
print(strong_changes)
```

```python
# Calculate the number of strong changes %%
num_strong_changes = strong_changes.shape[0]
total_locations = df.shape[0]

# Calculate percentages
percentage_increasing = (strong_changes[strong_changes['trend_bin'] == 'stark steigend'].shape[0] / total_locations) * 100
percentage_decreasing = (strong_changes[strong_changes['trend_bin'] == 'stark sinkend'].shape[0] / total_locations) * 100

print(f"Percentage of locations with strongly increasing groundwater levels: {percentage_increasing}%")
print(f"Percentage of locations with strongly decreasing groundwater levels: {percentage_decreasing}%")
```

```python
# Plotting the trend normalized by Behoerde
plt.figure(figsize=(12, 6))
sns.countplot(data=strong_changes, x='trend_bin', palette=['red', 'green'])
plt.title('Distribution of Strongly Changing Groundwater Levels')
plt.xlabel('Trend Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter for locations where trend_bin indicates a decrease
decreasing_trend = df[df['trend_bin'] == 'stark sinkend']

# Sort by 'trend_normalized' to get the largest decreases
top_decreasing = decreasing_trend.sort_values(by='trend_normalized').head(20)  # Change to head(10) for top 10

# Display the top 20 locations
top_decreasing
```

```python
# Plotting the top 20 locations with the most decrease in groundwater levels
plt.figure(figsize=(12, 6))
sns.barplot(data=top_decreasing, x='kreis', y='trend_normalized', palette='Reds')
plt.title('Top 20 Locations with the Most Decrease in Groundwater Levels')
plt.xlabel('District (Kreis)')
plt.ylabel('Normalized Trend (Decrease)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter for locations where trend_bin indicates a decrease
increasing_trend = df[df['trend_bin'] == 'stark steigend']

# Sort by 'trend_normalized' to get the largest decreases
top_increasing = increasing_trend.sort_values(by='trend_normalized').head(20)  # Change to head(10) for top 10

# Display the top 20 locations
top_increasing
```

```python
# Plotting the top 20 locations with the most increase in groundwater levels
plt.figure(figsize=(12, 6))
sns.barplot(data=top_increasing, x='kreis', y='trend_normalized', palette='Blues')
plt.title('Top 20 Locations with the Most Increase in Groundwater Levels')
plt.xlabel('District (Kreis)')
plt.ylabel('Normalized Trend (Decrease)')
plt.xticks(rotation=45)
plt.show()
```
